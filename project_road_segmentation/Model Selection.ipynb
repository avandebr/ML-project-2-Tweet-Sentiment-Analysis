```python
%matplotlib inline
import matplotlib.image as mpimg
import numpy as np
import matplotlib.pyplot as plt
import os,sys
from PIL import Image
from helpers import *
from feature_extraction import *
from sklearn import metrics
from sklearn import linear_model
from sklearn.cross_validation import cross_val_score
%load_ext autoreload
%autoreload 2
```

# 1 - Logistic Regression

### A - Does balancing the amount of sample from each class affect the F1 Score

```python
## --------------- Without balancing ----------------
c = []
accuracies = []
for x in np.logspace(-3,8,50):
    logreg = linear_model.LogisticRegression(C=x, class_weight="balanced")
    scores = cross_val_score(logreg, X_train, Y_train, cv=5, scoring='f1_macro')
    c.append(x)
    accuracies.append(scores.mean())
    #print("c = {:20.5f} Accuracy: {:10.9f}".format(x,scores.mean()))

plt.semilogx(c, accuracies)
plt.title('F1 Score for Linear classifier no balancing')
plt.xlabel('hyperparameter c')
plt.ylabel('mean F1 score')
plt.grid(True)
plt.savefig(ROOT_DIR+'F1ScoreNoBalancing.png')
```

```python
## --------------- With balancing ----------------
X_train_balanced, Y_train_balanced = balance_training_set(X_train,Y_train)
c = []
accuracies = []
for x in np.logspace(-3,8,50):
    logreg = linear_model.LogisticRegression(C=x, class_weight="balanced")
    scores = cross_val_score(logreg, X_train_balanced, Y_train_balanced, cv=5, scoring='f1_macro')
    c.append(x)
    accuracies.append(scores.mean())
    #print("c = {:20.5f} Accuracy: {:10.9f}".format(x,scores.mean()))

plt.semilogx(c, accuracies)
plt.title('F1 Score for Linear classifier with balancing')
plt.xlabel('hyperparameter c')
plt.ylabel('mean F1 score')
plt.grid(True)
plt.savefig(ROOT_DIR+'F1ScoreWithBalancing.png')
```

**Conclusion** : We see that the F1 score is much better when we actually balance the input data, also this allows us to decrease the amount of data to train on.

### Does the amount of training sample affect the F1 Score
We try to see if we need to reduce the amount of training samples or if we can keep it to be the total amount of data

```python
n_sample = len(X_train)
shuffled_indices = np.random.permutation(range(n_sample))
length_divisors = np.arange(1,15)
length = [n_sample//i for i in length_divisors]
max_accuracies = []
c_achiever = []
for l in length:
    X_train_short = X_train[shuffled_indices[:l]]
    Y_train_short = Y_train[shuffled_indices[:l]]
    c = []
    accuracies = []
    for x in np.logspace(-3,8,30):
        logreg = linear_model.LogisticRegression(C=x, class_weight="balanced")
        scores = cross_val_score(logreg, X_train_short, Y_train_short, cv=5, scoring='f1_macro')
        c.append(x)
        accuracies.append(scores.mean())
    max_accuracy = max(accuracies)
    max_accuracies.append(max_accuracy)
    index = np.argmax(max_accuracy)
    c_achiver = c[index]
        
plt.plot(length, max_accuracies)
plt.title('F1 Score for Linear classifier depending on the number of samples')
plt.xlabel('number of samples')
plt.ylabel('Maximum (mean) F1 score')
plt.grid(True)
plt.savefig(ROOT_DIR+'F1ScoreLength.png')
```

**Conclusion** : Mutiple run gives different results depending on which samples are considered, but the general trend remains that the more samples we have the better when doing 2 features dimension
